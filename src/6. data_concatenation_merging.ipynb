```python
import pandas as pd
```

```python
weather_india=  pd.DataFrame({
    "City": ['Gwalior','Delhi','Banglore'],
    "Temperature": [ 27, 16, 18],
    "Humidity": [ 60.7, 67, 69]
})
weather_india
```

```python
weather_nepal= pd.DataFrame({
    'City': [ 'Kathmandu', 'Bharatpur', 'Pokhara', 'Kirtipur'],
    'Temperature': [ 9, 17, 10, 10],
    'Humidity': [ 49, 34, 36, 50]
})
weather_nepal
```

```python
weather_india_nepal= pd.concat([weather_india, weather_nepal])
weather_india_nepal
```

```python
weather= pd.concat([weather_india, weather_nepal], ignore_index= True)
weather
```

```python
df= pd.concat([weather_india, weather_nepal])
df.index = range(1, len(df) + 1)
df
```

```python
df_key= pd.concat([weather_india, weather_nepal], keys= [ "India", "Nepal"])
df_key
```

```python
df_key.loc["Nepal"]
```

### Column-wise concatenation

```python
temperature_india=  pd.DataFrame({
    "City": ['Gwalior','Delhi'],
    "Temperature": [ 27, 16],
}, index= [0, 1])
temperature_india
```

```python
weather_india=  pd.DataFrame({
    "City": ['Delhi', 'Banglore'],
    "Humidity": [ 67, 69]
}, index= [ 1, 2])
weather_india
```

```python
weather_02= pd.concat([temperature_india,weather_india], axis= 1) 
weather_02
```

### Merge

```python
weather_merge= pd.merge( temperature_india, weather_india, on= "City")
weather_merge
```

```python
weather_merge_inner_join= pd.merge( temperature_india, weather_india, on= "City", how= 'inner')
weather_merge_inner_join
```

```python
weather_merge_left_join= pd.merge( temperature_india, weather_india, on= "City", how= 'left')
weather_merge_left_join
```

```python
weather_merge_right_join= pd.merge( temperature_india, weather_india, on= "City", how= 'right')
weather_merge_right_join
```

```python
weather_merge_outer_join= pd.merge( temperature_india, weather_india, on= "City", how= 'outer')
weather_merge_outer_join
```
